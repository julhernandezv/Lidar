## Import packages and $lidar$ module

The lidar module must had been instaled previously.

```python
import pandas as pd
import numpy as np
import datetime as dt
import os, locale
# os.system('rm lidar/lidar/*.pyc')
from lidar.lidar import Lidar
locale.setlocale(locale.LC_TIME, ('en_GB','utf-8'))
%matplotlib inline
# vlim = {'analog-s':[0.2,16],'analog-p':[0.2,16],'analog':[0,0.7] }
```

## Read datafiles

In $\textbf{date}$ you can choose initial and final dates for the analysis.

In $\textbf{binario}$ object the data selected will be stored. The kwg of this function are:

- fechaI
- fechaF
- scan ...
- output..
- path...


```python
# for date in pd.date_range('2018-06-01','2018-10-11',freq='d'): #'2018-06-27','2018-07-14',freq='d'):
#     try:
date = pd.date_range('2018-10-23','2018-10-23',freq='d')[0]
# date = pd.date_range('2018-06-30','2018-06-30',freq='d')[0] #'2018-06-27','2018-07-14',freq='d'):
        #
binario =   Lidar(
    fechaI=date.strftime('%Y-%m-%d'),
    fechaF=date.strftime('%Y-%m-%d'),
    scan='FixedPoint',
    # scan='3D',
    output='raw',
    path='CalidadAire/JupyerNotebooks/sjaramillo/Lidar_things/Figuras'
)
```

## Display Data and products

$\textbf{binario}$ object has many properties. For example, $\textbf{binario.datos}$ entrega todos los datos indexados por la fecha y con las columnas en tres niveles:
- Level 1: Height (km) from 0.00375 to the end.
- Level 2: Zenith angle (possible values between -5 and 90). In fixed point this value is constant.
- Level 3: Parameters (analos-p/s, photon p/s)
